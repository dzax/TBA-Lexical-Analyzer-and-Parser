## Parser

```python
#input example
sentence_parser = input()
tokens = sentence_parser.lower().split()
tokens.append('EOS')

#symbols definition
non_terminals = ['S', 'NN', 'VB']
terminals = ['awak', 'bermain', 'catur', 'karom', 'kereta', 'komputer', 'saya', 'motosikal', 'memandu', 'melihat']

#parser table definition
parse_table = {}

parse_table[('S', 'saya')] = ['NN', 'VB', 'NN']
parse_table[('S', 'awak')] = ['NN', 'VB', 'NN']
parse_table[('S', 'komputer')] = ['NN', 'VB', 'NN']
parse_table[('S', 'kereta')] = ['NN', 'VB', 'NN']
parse_table[('S', 'karom')] = ['NN', 'VB', 'NN']
parse_table[('S', 'catur')] = ['NN', 'VB', 'NN']
parse_table[('S', 'motosikal')] = ['NN', 'VB', 'NN']
parse_table[('S', 'bermain')] = ['error']
parse_table[('S', 'memandu')] = ['error']
parse_table[('S', 'melihat')] = ['error']
parse_table[('S', 'EOS')] = ['error']

parse_table[('NN', 'saya')] = ['saya']
parse_table[('NN', 'awak')] = ['awak']
parse_table[('NN', 'komputer')] = ['komputer']
parse_table[('NN', 'kereta')] = ['kereta']
parse_table[('NN', 'karom')] = ['karom']
parse_table[('NN', 'catur')] = ['catur']
parse_table[('NN', 'motosikal')] = ['motosikal']
parse_table[('NN', 'bermain')] = ['error']
parse_table[('NN', 'memandu')] = ['error']
parse_table[('NN', 'melihat')] = ['error']
parse_table[('NN', 'EOS')] = ['error']

parse_table[('VB', 'saya')] = ['error']
parse_table[('VB', 'awak')] = ['error']
parse_table[('VB', 'komputer')] = ['error']
parse_table[('VB', 'kereta')] = ['error']
parse_table[('VB', 'karom')] = ['error']
parse_table[('NN', 'catur')] = ['error']
parse_table[('VB', 'motosikal')] = ['error']
parse_table[('VB', 'bermain')] = ['bermain']
parse_table[('VB', 'memandu')] = ['memandu']
parse_table[('VB', 'melihat')] = ['melihat']
parse_table[('VB', 'EOS')] = ['error']

#stack initialization
stack = []
stack.append('#')
stack.append('S')

#input reading initialization
idx_token = 0
symbol = tokens[idx_token]

#parsing process
while (len(stack)>0):
  top = stack[len(stack)-1]
  print('top = ', top)
  print('symbol = ', symbol)
  if top in terminals:
    print('top stack is a terminal symbol')
    if top==symbol:
      stack.pop()
      idx_token = idx_token + 1
      symbol = tokens[idx_token]
      if symbol == 'EOS':
        print('isi stack:', stack)
        stack.pop()
    else:
      print('error')
      break;
  elif top in non_terminals:
    print('top stack is a non-terminal symbol')
    if parse_table[(top, symbol)][0] != 'error':
      stack.pop()
      symbols_to_be_pushed = parse_table[(top, symbol)]
      for i in range(len(symbols_to_be_pushed)-1,-1,-1):
        stack.append(symbols_to_be_pushed[i])
    else:
      print('error')
      break;
  else:
    print('error')
    break;
  print('stack elements:', stack)
  print()

#conclusion
print()
if symbol == 'EOS' and len(stack)==0:
  print('input string', sentence_parser, 'accepted, grammarly correct')
else:
  print('Error, input string:', sentence_parser, ', rejected')
```
